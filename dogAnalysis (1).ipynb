# UNIVARIATE ANALYSIS

```python
fig, axes = plt.subplots(5, 2, figsize = (25, 20))
sns.set_style('whitegrid')
fig.suptitle('Bar plot for all categorical variables in the dataset')

ax = sns.countplot(ax = axes[0, 0], x = 'Highest Price Sold', data = data, palette = 'RdBu_r', 
              order = data['Highest Price Sold'].value_counts().head(10).index);
ax = sns.countplot(ax = axes[0, 1], x = 'Category', data = data, palette = 'rainbow',
             order = data['Category'].value_counts().head(10).index);
ax = sns.countplot(ax = axes[1, 0], x = 'Lowest Price Sold', data = data, palette="Set1",
             order = data['Lowest Price Sold'].value_counts().head(10).index);
ax = sns.countplot(ax = axes[1, 1], x = 'LifeTime Cost', data = data, palette="PuOr", 
              order = data['LifeTime Cost'].value_counts().head(10).index);
ax = sns.countplot(ax = axes[2, 0], x = 'Trainability', data = data, color = 'blue',
             order = data['Trainability'].value_counts().head(10).index);
ax =sns.countplot(ax = axes[2, 1], x = 'Longevity', data = data, palette="RdBu",
             order = data['Longevity'].value_counts().head(10).index);
ax =sns.countplot(ax = axes[3, 0], x = 'Avg Price(USA)', data = data, color = 'red', 
              order = data['Avg Price(USA)'].value_counts().head(10).index);
ax =sns.countplot(ax = axes[3, 1], x = 'Sustainability % For Children', data = data, palette="Set2",
             order = data['Sustainability % For Children'].value_counts().head(10).index);
ax = sns.countplot(ax = axes[4, 0], x = 'Grooming Required', data = data, palette = 'rainbow', 
              order = data['Grooming Required'].value_counts().head(10).index);
ax = sns.countplot(ax = axes[4, 1], x = 'Intelligence', data = data, palette="Set2",
             order = data['Intelligence'].value_counts().head(10).index);

plt.rc('font', size=8)
plt.rc('axes', titlesize=100)
plt.rc('axes', labelsize=20)
plt.rc('xtick', labelsize=80)
plt.rc('ytick', labelsize=80)
plt.rc('legend', fontsize=80)
plt.rc('figure', titlesize=100)
```

# BiVariate Analysis

```python
figsize = (18, 20)
sns.set_style('whitegrid')
ax = sns.countplot(y='Intelligence', hue = 'Size', data=data, palette='colorblind', order = data['Intelligence'].value_counts().head().index)
total = data['Intelligence'].count()
ax.bar_label(ax.containers[0], fmt=lambda x: f'{(x/total)*100:.1f}%')
ax.margins(x=0.1)
ax.set(xlabel='Count', ylabel='Intelligence ', title='Intelligence  by Size ')
ax.set_xticklabels(ax.get_xticklabels(), fontsize=7)
plt.tight_layout()
plt.show()

plt.rc('font', size=8)
plt.rc('axes', titlesize=10)
plt.rc('axes', labelsize=10)
plt.rc('xtick', labelsize=18)
plt.rc('ytick', labelsize=18)
plt.rc('legend', fontsize=15)
plt.rc('figure', titlesize=10)
```
